```python
!pip install opencv-python-headless scikit-image scikit-learn

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
import torchvision.models as models
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive, files
import os
from PIL import Image
from skimage.feature import hog
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import pickle

drive.mount('/content/drive')

!mkdir -p /content/drive/MyDrive/worker_marketplace/signature_recognition

class SignatureExtractor:
    """Class to extract signatures from documents using image processing techniques"""

    def __init__(self):
        pass

    def preprocess(self, image):
        """Preprocess image for signature extraction"""
        if isinstance(image, str):
            image = cv2.imread(image)

        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

        return thresh

    def find_signature_areas(self, thresh_img, min_area=1000, max_area=50000):
        """Find potential signature areas"""
        contours, _ = cv2.findContours(thresh_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        signature_regions = []
        for contour in contours:
            area = cv2.contourArea(contour)
            if min_area < area < max_area:
                x, y, w, h = cv2.boundingRect(contour)
                signature_regions.append((x, y, w, h))

        return signature_regions

    def extract_signature(self, image_path, visualize=False):
        """Extract the signature from a document"""
        image = cv2.imread(image_path)
        original = image.copy()

        thresh = self.preprocess(image)

        regions = self.find_signature_areas(thresh)

        if len(regions) == 0:
            return None

        regions.sort(key=lambda r: r[2] * r[3], reverse=True)

        x, y, w, h = regions[0]
        signature = original[y:y+h, x:x+w]

        if visualize:
            cv2.rectangle(original, (x, y), (x+w, y+h), (0, 255, 0), 2)

            plt.figure(figsize=(15, 10))
            plt.subplot(1, 2, 1)
            plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
            plt.title("Detected Signature")

            plt.subplot(1, 2, 2)
            plt.imshow(cv2.cvtColor(signature, cv2.COLOR_BGR2RGB))
            plt.title("Extracted Signature")
            plt.show()

        return signature

class SignatureDataset(Dataset):
    """Dataset for signature verification"""

    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.genuine_dir = os.path.join(root_dir, "genuine")
        self.forged_dir = os.path.join(root_dir, "forged")

        self.samples = []
        self.labels = []

        for person_dir in os.listdir(self.genuine_dir):
            person_path = os.path.join(self.genuine_dir, person_dir)
            if not os.path.isdir(person_path):
                continue

            signature_files = [f for f in os.listdir(person_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

            for i, file1 in enumerate(signature_files):
                for j, file2 in enumerate(signature_files):
                    if i != j:
                        self.samples.append((
                            os.path.join(person_path, file1),
                            os.path.join(person_path, file2)
                        ))
                        self.labels.append(1)

                forged_path = os.path.join(self.forged_dir, person_dir)
                if os.path.exists(forged_path):
                    forged_files = [f for f in os.listdir(forged_path) if f.endswith(('.png', '.jpg', '.jpeg'))]
                    for forged_file in forged_files:
                        self.samples.append((
                            os.path.join(person_path, file1),
                            os.path.join(forged_path, forged_file)
                        ))
                        self.labels.append(0)

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path1, img_path2 = self.samples[idx]
        img1 = Image.open(img_path1).convert('L')
        img2 = Image.open(img_path2).convert('L')

        if self.transform:
            img1 = self.transform(img1)
            img2 = self.transform(img2)

        return img1, img2, self.labels[idx]

class SiameseNetwork(nn.Module):
    """Siamese neural network for signature verification"""

    def __init__(self):
        super(SiameseNetwork, self).__init__()

        self.feature_extractor = models.resnet18(pretrained=True)
        self.feature_extractor.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
        self.feature_extractor = nn.Sequential(*list(self.feature_extractor.children())[:-1])

        self.embedding = nn.Sequential(
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, 128)
        )

    def forward_one(self, x):
        x = self.feature_extractor(x)
        x = x.view(x.size(0), -1)
        x = self.embedding(x)
        return x

    def forward(self, input1, input2):
        output1 = self.forward_one(input1)
        output2 = self.forward_one(input2)
        return output1, output2

class SignatureVerifier:
    """Class for signature verification"""

    def __init__(self, model=None, threshold=0.7):
        self.model = model
        self.threshold = threshold
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.5], std=[0.5])
        ])

    def load_model(self, model_path):
        """Load a trained model"""
        self.model = SiameseNetwork()
        self.model.load_state_dict(torch.load(model_path))
        self.model.eval()

    def verify(self, sig1_path, sig2_path):
        """Verify if two signatures match"""
        if self.model is None:
            raise ValueError("Model not loaded. Call load_model first.")

        img1 = Image.open(sig1_path).convert('L')
        img2 = Image.open(sig2_path).convert('L')

        img1 = self.transform(img1).unsqueeze(0)
        img2 = self.transform(img2).unsqueeze(0)

        with torch.no_grad():
            output1, output2 = self.model(img1, img2)

            cosine = nn.CosineSimilarity(dim=1)
            similarity = cosine(output1, output2).item()

        return similarity > self.threshold, similarity

def extract_hog_features(image_path):
    """Extract HOG features from an image"""
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (128, 64))

    features, _ = hog(
        img,
        orientations=9,
        pixels_per_cell=(8, 8),
        cells_per_block=(2, 2),
        visualize=True,
        block_norm='L2-Hys'
    )

    return features

def train_svm_model(dataset_path):
    X = []
    y = []

    genuine_dir = os.path.join(dataset_path, "genuine")

    for person_dir in os.listdir(genuine_dir):
        person_path = os.path.join(genuine_dir, person_dir)
        if not os.path.isdir(person_path):
            continue

        signature_files = [f for f in os.listdir(person_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

        for sig_file in signature_files:
            sig_path = os.path.join(person_path, sig_file)
            features = extract_hog_features(sig_path)
            X.append(features)
            y.append(person_dir)

    classifier = SVC(kernel='rbf', probability=True)
    classifier.fit(X, y)

    with open('/content/drive/MyDrive/worker_marketplace/signature_recognition/svm_model.pkl', 'wb') as f:
        pickle.dump(classifier, f)

    return classifier

def demo_signature_recognition():
    extractor = SignatureExtractor()

    print("Upload a document with a signature:")
    uploaded = files.upload()

    if len(uploaded) > 0:
        doc_path = list(uploaded.keys())[0]

        signature = extractor.extract_signature(doc_path, visualize=True)

        if signature is not None:
            cv2.imwrite('extracted_signature.jpg', signature)
            print("Signature successfully extracted and saved as 'extracted_signature.jpg'")

            print("Upload a known signature for verification:")
            known_sig = files.upload()

            if len(known_sig) > 0:
                known_path = list(known_sig.keys())[0]

                known_features = extract_hog_features(known_path)
                extracted_features = extract_hog_features('extracted_signature.jpg')

                similarity = np.dot(known_features, extracted_features) / (np.linalg.norm(known_features) * np.linalg.norm(extracted_features))

                print(f"Signature similarity: {similarity:.4f}")
                print(f"Verification result: {'Match' if similarity > 0.7 else 'No match'}")

                fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
                ax1.imshow(cv2.cvtColor(cv2.imread('extracted_signature.jpg'), cv2.COLOR_BGR2RGB))
                ax1.set_title("Extracted Signature")
                ax1.axis('off')
                ax2.imshow(cv2.cvtColor(cv2.imread(known_path), cv2.COLOR_BGR2RGB))
                ax2.set_title("Known Signature")
                ax2.axis('off')
                plt.suptitle(f"Similarity: {similarity:.4f}")
                plt.show()
        else:
            print("No signature detected in the document.")
    else:
        print("No document uploaded.")

demo_signature_recognition()

with open('/content/drive/MyDrive/worker_marketplace/signature_recognition/signature_extractor.pkl', 'wb') as f:
    pickle.dump(SignatureExtractor(), f)

print("Signature recognition system saved to Google Drive")
```
